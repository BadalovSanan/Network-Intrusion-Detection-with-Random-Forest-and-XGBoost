```python
fig, ax = plt.subplots(figsize=(12, 6))

counts = df['Label'].value_counts()
colors = plt.cm.Blues(np.linspace(0.4, 0.8, len(counts)))
bars = ax.bar(counts.index, counts.values, color=colors, edgecolor='none')

ax.set_yscale('log')
ax.bar_label(bars, fmt='{:,}', padding=5, fontsize=10, fontweight='bold')

plt.title('Frequency of Each Label', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Labels', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, ha='right')

for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.savefig('../results/figures/frequency.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(20, 16))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.savefig('../results/figures/first_heatmap.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
corr = numeric_df.corr()

# unstack turns the matrix into pairs, then we filter and sort
corr_pairs = corr.unstack()
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]  # remove self-correlation (feature vs itself = always 1.0)
corr_pairs = corr_pairs.drop_duplicates()  # each pair appears twice (A-B and B-A), keep one

sorted_pairs = corr_pairs.abs().sort_values(ascending=False)
```
