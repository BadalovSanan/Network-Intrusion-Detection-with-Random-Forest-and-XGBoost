```python
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
```

```python
X_test = pd.read_csv('../data/processed/X_test.csv')
y_test = pd.read_csv('../data/processed/y_test.csv').values.ravel()
```

```python
best_rf = joblib.load('../models/random_forest_best.pkl')
best_xgb = joblib.load('../models/xgb_best.pkl')
```

```python
preds_rf = best_rf.predict(X_test)
preds_xgb = best_xgb.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, preds_rf)
f1_rf = f1_score(y_test, preds_rf, average='macro')

print("-" * 20 + " Random Forest Results " + "-" * 20)
print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, preds_rf))
print("-" * 51)
print('\nClassification Report:\n')
print(classification_report(y_test, preds_rf))
print("-" * 51)
print(f'\nAccuracy Score: {accuracy_rf:.4f}')
print("-" * 51)
print(f'\nF1 Score (Macro): {f1_rf:.4f}')
```

```python
accuracy_xgb = accuracy_score(y_test, preds_xgb)
f1_xgb = f1_score(y_test, preds_xgb, average='macro')

print("-" * 20 + " XGBoost Results " + "-" * 20)
print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, preds_xgb))
print("-" * 51)
print('\nClassification Report:\n')
print(classification_report(y_test, preds_xgb))
print("-" * 51)
print(f'\nAccuracy Score: {accuracy_xgb:.4f}')
print("-" * 51)
print(f'\nF1 Score (Macro): {f1_xgb:.4f}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 10))

sns.heatmap(confusion_matrix(y_test, preds_rf), annot=True, fmt='d', cmap='Blues', ax=ax1)
ax1.set_title('Random Forest Confusion Matrix', fontsize=14, fontweight='bold')
ax1.set_xlabel('Predicted Labels', fontsize=12)
ax1.set_ylabel('True Labels', fontsize=12)

sns.heatmap(confusion_matrix(y_test, preds_xgb), annot=True, fmt='d', cmap='Oranges', ax=ax2)
ax2.set_title('XGBoost Confusion Matrix', fontsize=14, fontweight='bold')
ax2.set_xlabel('Predicted Labels', fontsize=12)
ax2.set_ylabel('True Labels', fontsize=12)

plt.tight_layout()
plt.savefig('../results/figures/evaluation_of_results.png', dpi=200, bbox_inches='tight')
plt.show()
```
